```python
from pyspark import SparkConf, SparkContext
import sys
```

```python
def binary_ratings(data):
    #fields = data.split(",")
    new_line = data.map(lambda x: x[2])
    star_rating = 1 if data[2] else 0
    #print(f"{data[0]}, {data[1]}, {star_rating}")
    print(new_line.take(5))
    return f"{data[0]}, {data[1]}, {star_rating}"



```

```python
def hash_function(num, bins):
    return (2*(num) + 10) % bins
```

```python
def jaccard_similarity(business1, business2):
    intersection = business1.intersection(business2)
    union = business1.union(business2)
    return len(intersection) / len(union)
```

```python
def LSH(data):
    ratings = []
    for items in data:
        similarity = jaccard_similarity(items[0], items[1])
        if similarity >= 0.5:
            ratings.extend(items)
```

```python
def main(input_path, output_path):
    conf = SparkConf().setAppName("task1")
    sc = SparkContext(conf=conf)
    sc.setLogLevel("WARN")

    raw_data = sc.textFile(input_path)
    print(raw_data.take(10))

    data = binary_ratings(raw_data)
    LSH(data)

```

```python
if __name__ == '__main__':
    input_path = sys.argv[1]
    output_path = sys.argv[2]

    main(input_path, output_path)

```
